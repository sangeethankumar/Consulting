Creating Camera Instance

```python
import pypylon.pylon as py
import os
```

```python
os.environ["PYLON_CAMEMU"] = "1"
```

```python
tlf = py.TlFactory.GetInstance()
devices = tlf.EnumerateDevices()
device = devices[0]
```

```python
device
```

```python
cd = tlf.CreateDevice(device)
cam = py.InstantCamera(cd)
cam.Open()
```

```python
cam.TriggerSelector.Symbolics
```

```python
fw = 1280
fh = 1024
fps = 200
```

```python
cam.Width.Value = fw
cam.Height.Value = fh
cam.AcquisitionFrameRateEnable.Value = True
cam.AcquisitionFrameRate.Value = fps
```

```python
res = cam.GrabOne(1000)
```

```python
res.GetTimeStamp()
```

```python
cam.Close()
```
